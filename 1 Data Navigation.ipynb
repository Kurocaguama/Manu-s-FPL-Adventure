**Filtrado de datos**

Por el momento se tienen dos metodologías de filtración de datos, una por equipo (puede ser buena por si ocupas un sleeper pick de un equipo en particular), o por jugador (creo que es lo más relevante para FPL).

```python
# Idea one for info filtering:
# Team -> Player

team_split = df.groupby('Team')
arsenal_df = team_split.get_group('Arsenal')
arsenal_df = arsenal_df.groupby('Player')
kai = arsenal_df.get_group('Kai Havertz')
kai.head()
```

```python
# Idea two for filtering:
# Player only
player_split = df.groupby('Player')
bruno = player_split.get_group('Bruno Fernandes')
bruno.head()
```

```python
def get_player_info(player_name):
    """
        player_name = str ; Debe coincidir en su totalidad con el nombre del jugador
    """
    player_split = df.groupby('Player')
    player_df = player_split.get_group(player_name)
    mins = round(player_df['Minutes'].mean(), 2)
    goals = round(player_df['Goals'].mean(), 2)
    passes = round(player_df['Passes Completed'].mean(), 2)
    att_passes = round(player_df['Passes Attempted'].mean(), 2)
    pass_complete_ratio = round(passes/att_passes, 2)

    print('==============================')
    print('==========Bienvenido==========')
    print('==============================')
    print('Jugador elegido: {}'.format(player_name))
    print('Promedio de minutos: {}'.format(mins))
    print('Goles por partido: {}'.format(goals))
    print('Porcentaje de pases completos: {}%'.format(pass_complete_ratio))
    return mins, goals, pass_complete_ratio
```

```python
get_player_info('Kevin De Bruyne')
```

```python
def get_player_info_aux(player_name):
    """
        player_name = str ; Debe coincidir en su totalidad con el nombre del jugador
    """
    player_split = df.groupby('Player')
    player_df = player_split.get_group(player_name)
    mins = round(player_df['Minutes'].mean(), 2)
    goals = round(player_df['Goals'].mean(), 2)
    passes = round(player_df['Passes Completed'].mean(), 2)
    att_passes = round(player_df['Passes Attempted'].mean(), 2)
    pass_complete_ratio = round(passes/att_passes, 2)
    return mins, goals, pass_complete_ratio

def compare_players(player1, player2):
    p1_mins, p1_goals, p1_ratio = get_player_info_aux(player1)
    p2_mins, p2_goals, p2_ratio = get_player_info_aux(player2)
    print('===============================')
    print('=========HEAD TO HEAD==========')
    print('===============================')
    print('Jugadores a comparar: {} vs {}'.format(player1, player2))
    print('-------')
    print('Promedio de minutos por partido')
    print(f'{player1}: {p1_mins}')
    print(f'{player2}: {p2_mins}')
    print('-------')
    print('Goles por partido')
    print(f'{player1}: {p1_goals}')
    print(f'{player2}: {p2_goals}')
    print('-------')
    print('Porcentaje de pases completos')
    print(f'{player1}: {p1_ratio}')
    print(f'{player2}: {p1_ratio}')

compare_players('Bruno Fernandes', 'Kevin De Bruyne')
```
